```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import statistics
```

```python
data = pd.read_csv("diabetes_risk_prediction_dataset.csv")
data.head()
data=data.replace('Yes',1)
data=data.replace('No',0)
data=data.replace('Positive',1)
data=data.replace('Negative',0)
data=data.replace('Male',1)
data=data.replace('Female',0)
```

```python
X=data['Age']
classes=data['class']
X_train, X_test, c_train, c_test = train_test_split(X,classes ,random_state=104, test_size=0.2, shuffle=True)
plt.scatter(X, X, c=classes)
```

```python
tr_data = list(zip(X_train, X_train))

knn = KNeighborsClassifier(3)

knn.fit(tr_data, c_train)
```

```python
te_data = list(zip(X_test, X_test))
rng = len(te_data)
te_res=[]
i=0
while i < rng:
  new_point = [te_data[i]]
  prediction = knn.predict(new_point)
  te_res.append(prediction)
  i=i+1

plt.scatter(x=X_test, y=X_test, c=te_res)
print(te_res[0][0])
```

```python
FalsePos=0
FalseNeg=0
TruePos=0
TrueNeg=0
i=0
while i<rng:
  if c_test.values[i]==te_res[i][0] and c_test.values[i]==0:
    TrueNeg=TrueNeg+1
  if c_test.values[i]==te_res[i][0] and c_test.values[i]==1:
    TruePos=TruePos+1
  if c_test.values[i]!=te_res[i][0] and c_test.values[i]==0:
    FalsePos=FalsePos+1
  if c_test.values[i]!=te_res[i][0] and c_test.values[i]==1:
    FalseNeg=FalseNeg+1
  i=i+1
print(FalsePos,FalseNeg,TruePos,TrueNeg)
```
