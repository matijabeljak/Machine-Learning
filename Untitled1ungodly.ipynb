```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics

data = pd.read_csv("Real estate.csv")
data.head()






sns.scatterplot(data, x=data['X1 transaction date'], y=data['Y house price of unit area'])
plt.show()
sns.scatterplot(data, x=data['X2 house age'], y=data['Y house price of unit area'])
plt.show()
a=statistics.stdev(data['Y house price of unit area'])
dat=data['Y house price of unit area']
step=(max(dat)-min(dat))//10
bat=[][len(dat)]

for i in range(len(dat)):
  dat[i]
  if dat[i]<min(dat)+step:
    bat[1][dat[i]].append
  if dat[i]<min(dat)+2*step:
    bat[2][dat[i]].append
  if dat[i]<min(dat)+3*step:
    bat[3][dat[i]].append
  if dat[i]<min(dat)+4*step:
    bat[4][dat[i]].append
  if dat[i]<min(dat)+5*step:
    bat[5][dat[i]].append
  if dat[i]<min(dat)+6*step:
    bat[6][dat[i]].append
  if dat[i]<min(dat)+7*step:
    bat[7][dat[i]].append
  if dat[i]<min(dat)+8*step:
    bat[8][dat[i]].append
  if dat[i]<min(dat)+9*step:
    bat[9][dat[i]]
  if dat[i]<min(dat)+10*step:
    bat[10][dat[i]]








print(a)
sns.scatterplot(data, x=bat[i], y=sum(bat[i][]))
plt.show()
sns.scatterplot(data, x=data['X4 number of convenience stores'], y=data['Y house price of unit area'])
sns.lineplot()
plt.show()
sns.scatterplot(data, x=data['X5 latitude'], y=data['X6 longitude'], hue=data['Y house price of unit area'], size=data['X4 number of convenience stores'])
plt.show()
sns.scatterplot(data, x=data['X6 longitude'], y=data['Y house price of unit area'])
plt.show()


cor = data.corr()
sns.heatmap(cor,annot=True,cmap="vlag",center=0)
```
